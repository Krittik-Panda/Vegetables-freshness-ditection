```python
import joblib
from features import extract_features

model = joblib.load("lightgbm_vegetable_freshness.pkl")
print("Model Loaded")
```

```python
import pandas as pd

def predict_image(image_path):

    features = extract_features(image_path)
    features = features.reshape(1, -1)

    # Convert to DataFrame with SAME feature names
    features_df = pd.DataFrame(features, columns=model.feature_name_)

    prediction = model.predict(features_df)[0]
    probability = model.predict_proba(features_df)[0]

    label = "Rotten" if prediction == 1 else "Fresh"

    print("Prediction:", label)
    print("Probability:", probability)
```

```python
predict_image("fresh_potato.jpg")
```
